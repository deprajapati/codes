```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = cv2.imread(r"C:\Users\Acer\Desktop\pet.jpg")
print(a)
```

```python
plt.imshow(a)
```

```python
b = cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
plt.imshow(b,cmap='gray')
```

# resize by-resolution

```python
print(a.shape)
br = cv2.resize(b,(1500, 800 ))
plt.imshow(br)
```

```python
cscaled = cv2.resize(b, None, fx=0.5, fy=0.5)
plt.imshow(cscaled)
```

```python
cblur = cv2.GaussianBlur(a,(19,19), 0)
plt.imshow(cblur)
cv2.imwrite('blured_pet.jpg',cblur)
```

```python
cgray = cv2.cvtColor(a,cv2.COLOR_RGB2GRAY)
cgray = cv2.GaussianBlur(cgray, (5,5), 0)
cgray = cv2.Canny(cgray,10,150)
plt.imshow(cgray, cmap='gray')
```

```python
cinvert = cv2.bitwise_not(cgray)
plt.imshow(cinvert, cmap='gray')
```
